```python
import pandas as pd
import numpy as np
```

```python
import json
```

```python
with open('subset.json', encoding='utf8') as f:
    data = json.load(f)
```

```python
data
```

```python
d = pd.DataFrame(data['data'])
```

```python
d.reset_index(inplace=True)
rows = []
_ = d.apply(lambda row: [rows.append([row.Country_Code, row.Sub_Region_Code, row.geo_index, nn])
                         for nn in row.Profile_urls], axis=1)
df_new = pd.DataFrame(rows, columns=d.columns[1:])
```

```python
df_new
```

```python
df_new.to_csv('data.csv')
```
